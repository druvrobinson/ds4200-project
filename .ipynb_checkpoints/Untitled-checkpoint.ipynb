```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = "/Users/sailibekov/Desktop/Classes/'24-'25 Classes/2nd sem/DS4200/project/StudentPerformanceFactors.csv"
df = pd.read_csv(data)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Hours_Studied'], df['Exam_Score'])
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship Between Hours Studied and Exam Score')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8q))
sns.boxplot(data=df, x="Family_Income", y="Exam_Score")
plt.xlabel("Family Income Level")
plt.ylabel("Exam Score")
plt.title("Distribution of Exam Scores by Family Income Level")
plt.show()
```
